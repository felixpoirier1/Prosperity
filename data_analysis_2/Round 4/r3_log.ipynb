```python
import pandas as pd
```

```python
l = []
with open("../../logs/64.csv", "r") as f:
    file = f.read().split(r"\n")
    for line in file:
        l.append(line.split(";"))
    #file = file.replace(r"\\n", "\n")
```

```python
df = pd.DataFrame(l[1:], columns = l[0])
```

```python
float_cols = df.columns[3:]
#df[[df.columns]]
df[float_cols] = df[float_cols].replace("", None).astype(float)
df[df.columns[:2]] = df[df.columns[:2]].astype(int)
```

```python
df.to_csv("../../logs/round3_results.csv", index=False)
```
