### Import Statements

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load Data

The codes below separately load all csv files as a dataframes, and expand columns using the ; separator. A column "new_timestamp" has been added to distinguish timestamps for each day when combined. A column 'day' has also been added to all 3 trades dataframes. 

```python
prices_0 = pd.read_csv("prices_round_3_day_0.csv", sep=";")
prices_0['new_timestamp'] = prices_0['timestamp']
prices_0
```

```python
prices_1 = pd.read_csv("prices_round_3_day_1.csv", sep=";")
prices_1['new_timestamp'] = prices_1['timestamp'] + 1000000
prices_1
```

```python
prices_2 = pd.read_csv("prices_round_3_day_2.csv", sep=";")
prices_2['new_timestamp'] = prices_2['timestamp'] + 2000000
prices_2
```

```python
trades_0 = pd.read_csv('trades_round_3_day_0_nn.csv', sep=";")
trades_0['new_timestamp'] = trades_0['timestamp']
trades_0.insert(0, 'day', '0')
trades_0
```

```python
trades_1 = pd.read_csv('trades_round_3_day_1_nn.csv', sep=";")
trades_1['new_timestamp'] = trades_1['timestamp'] + 1000000
trades_1.insert(0, 'day', '1')
trades_1
```

```python
trades_2 = pd.read_csv('trades_round_3_day_2_nn.csv', sep=";")
trades_2['new_timestamp'] = trades_2['timestamp'] + 2000000
trades_2.insert(0, 'day', '2')
trades_2
```

### Organize Data

The codes below first concatinate all 3 dataframes for price and trades into a single dataframe and then create separate dataframes for strawberries, chocolates, roses, and giftbasket respectively. A new column called spread is also created to find the spread between the bid and ask price. Indexes are also reset. Additionally, columns are converted from object to numeric data types. Finally, the prices dataframe is pivoted to provide the mid_price and spread for each product for every timestamp using the new_timestamp column. 

```python
prices = pd.concat([prices_0, prices_1, prices_2], ignore_index=True)

prices_columns_to_convert = ['day', 'timestamp', 'bid_price_1', 'bid_volume_1', 'bid_price_2', 'bid_volume_2', 'bid_price_3',
          'bid_volume_3', 'ask_price_1', 'ask_volume_1', 'ask_price_2', 'ask_volume_2', 'ask_price_3', 'ask_volume_3',
          'mid_price', 'profit_and_loss', 'new_timestamp']
for col in prices_columns_to_convert:
    prices[col] = pd.to_numeric(prices[col], errors='coerce')

prices['spread'] = prices['ask_price_1'] - prices['bid_price_1']

prices
```

```python
strawberry_prices = prices[prices['product'] == 'STRAWBERRIES']
strawberry_prices.reset_index(drop=True, inplace=True)
strawberry_prices
```

```python
chocolate_prices = prices[prices['product'] == 'CHOCOLATE']
chocolate_prices.reset_index(drop=True, inplace=True)
chocolate_prices
```

```python
rose_prices = prices[prices['product'] == 'ROSES']
rose_prices.reset_index(drop=True, inplace=True)
rose_prices
```

```python
gift_basket_prices = prices[prices['product'] == 'GIFT_BASKET']
gift_basket_prices.reset_index(drop=True, inplace=True)
gift_basket_prices
```

```python
trades = pd.concat([trades_0, trades_1, trades_2], ignore_index=True)

trades_columns_to_convert = ['day', 'timestamp', 'price', 'quantity']
for col in trades_columns_to_convert:
    trades[col] = pd.to_numeric(trades[col], errors='coerce')
    
trades
```

```python
strawberry_trades = trades[trades['symbol'] == 'STRAWBERRIES']
strawberry_trades.reset_index(drop=True, inplace=True)
strawberry_trades
```

```python
chocolate_trades = trades[trades['symbol'] == 'CHOCOLATE']
chocolate_trades.reset_index(drop=True, inplace=True)
chocolate_trades
```

```python
rose_trades = trades[trades['symbol'] == 'ROSES']
rose_trades.reset_index(drop=True, inplace=True)
rose_trades
```

```python
gift_basket_trades = trades[trades['symbol'] == 'GIFT_BASKET']
gift_basket_trades.reset_index(drop=True, inplace=True)
gift_basket_trades
```

```python
etf = prices.pivot(index='new_timestamp', columns='product', values=['mid_price', 'spread'])

etf.columns = ['_'.join(col) for col in etf.columns]

etf.rename(columns={
    'mid_price_CHOCOLATE': 'ch_price',
    'mid_price_ROSES': 'ro_price',
    'mid_price_GIFT_BASKET': 'gb_price',
    'mid_price_STRAWBERRIES': 'sb_price',
    'spread_CHOCOLATE': 'ch_spread',
    'spread_ROSES': 'ro_spread',
    'spread_GIFT_BASKET': 'gb_spread',
    'spread_STRAWBERRIES': 'sb_spread'
}, inplace=True)

etf.reset_index(inplace=True)

etf['combined_price'] = 6*etf['sb_price'] + 4*etf['ch_price'] + etf['ro_price']
etf['combined_price_wintercept'] = etf['combined_price'] + 400

etf
```

### Visualize Data

The codes below plot subplots for the mid_price for all 4 products. The mean values have also been highlighted. Additionally, the correlation matrix for mid_prices is also shown, showing that while all other prooducts are correlated very highlighly, strawberries are not.

```python
fig, axs = plt.subplots(4, 1, figsize=(8, 12))

# Plot each product's mid-price over time
axs[0].plot(etf['new_timestamp'], etf['sb_price'], label='Strawberries', color='red')
axs[0].axhline(y=etf['sb_price'].mean(), color='red', linestyle='--', label='Mean Price')
axs[0].fill_between(etf["new_timestamp"], etf['sb_price'].mean(), etf["sb_price"], where=(etf["sb_price"] <= etf['sb_price'].mean()), color='grey', alpha=0.3)
axs[0].set_ylabel('Mid-price')
axs[0].set_title('Strawberries Mid-price')

axs[1].plot(etf['new_timestamp'], etf['ch_price'], label='Chocolate', color='brown')
axs[1].axhline(y=etf['ch_price'].mean(), color='brown', linestyle='--', label='Mean Price')
axs[1].fill_between(etf["new_timestamp"], etf['ch_price'].mean(), etf["ch_price"], where=(etf["ch_price"] <= etf['ch_price'].mean()), color='grey', alpha=0.3)
axs[1].set_ylabel('Mid-price')
axs[1].set_title('Chocolate Mid-price')

axs[2].plot(etf['new_timestamp'], etf['ro_price'], label='Roses', color='pink')
axs[2].axhline(y=etf['ro_price'].mean(), color='pink', linestyle='--', label='Mean Price')
axs[2].fill_between(etf["new_timestamp"], etf['ro_price'].mean(), etf["ro_price"], where=(etf["ro_price"] <= etf['ro_price'].mean()), color='grey', alpha=0.3)
axs[2].set_ylabel('Mid-price')
axs[2].set_title('Roses Mid-price')

axs[3].plot(etf['new_timestamp'], etf['gb_price'], label='Gift Basket', color='green')
axs[3].axhline(y=etf['gb_price'].mean(), color='green', linestyle='--', label='Mean Price')
axs[3].fill_between(etf["new_timestamp"], etf['gb_price'].mean(), etf["gb_price"], where=(etf["gb_price"] <= etf['gb_price'].mean()), color='grey', alpha=0.3)
axs[3].set_xlabel('Timestamp')
axs[3].set_ylabel('Mid-price')
axs[3].set_title('Gift Basket Mid-price')

for ax in axs:
        ax.axvline(0, color='black', linestyle='--')
        ax.axvline(1000000, color='black', linestyle='--')
        ax.axvline(2000000, color='black', linestyle='--')
        ax.axvline(3000000, color='black', linestyle='--')

for ax in axs:
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(etf['new_timestamp'], etf['gb_price'], label='Gift Basket Price', color='green')
plt.plot(etf['new_timestamp'], etf['combined_price'], label='Combined Products Price', color='orange')

plt.axvline(0, color='black', linestyle='--')
plt.axvline(1000000, color='black', linestyle='--')
plt.axvline(2000000, color='black', linestyle='--')
plt.axvline(3000000, color='black', linestyle='--')

plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.title('Gift Basket Price vs Combined Products Price over Time')
plt.legend()
plt.grid(True)
plt.show()
```

The individual products combined into a basket tend to be priced lower than the gift basket package.

```python
price_columns = ['ch_price', 'ro_price', 'gb_price', 'sb_price', 'combined_price', 'combined_price_wintercept']

price_correlation_matrix = etf[price_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(price_correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, annot_kws={"size": 8})
plt.title('Correlation Matrix (Price Columns)')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Regression Analysis

```python
import statsmodels.api as sm

# Define the independent variable (X) and the dependent variable (y)
X = etf['combined_price']  # Independent variable (combined price)
y = etf['gb_price']        # Dependent variable (gift basket price)

# Add a constant term to the independent variable
X = sm.add_constant(X)

# Fit the OLS model
model = sm.OLS(y, X).fit()

# Print the summary of the regression results
print(model.summary())
```

```python
# Define the independent variable (X) and the dependent variable (y)
X2 = etf['combined_price']  # Independent variable (combined price)
y2 = etf['gb_price']        # Dependent variable (gift basket price)

# Fit the OLS model without intercept
model2 = sm.OLS(y2, X2).fit()

# Print the summary of the regression results
print(model2.summary())
```

```python
# Create lagged variables
etf_lag = etf.copy()

etf_lag['combined_lag1'] = etf_lag['combined_price'].shift(1)
etf_lag['combined_lag2'] = etf_lag['combined_price'].shift(10)
etf_lag['combined_lag3'] = etf_lag['combined_price'].shift(50)
etf_lag['combined_lag4'] = etf_lag['combined_price'].shift(100)

# Drop rows with NaN values resulting from the shift operation
etf_lag = etf_lag.dropna()
```

```python
# Define independent variables (lagged predictors) and the dependent variable
X3 = etf_lag['combined_lag3']
y3 = etf_lag['gb_price']

# Fit the regression model without intercept
model3 = sm.OLS(y3, X3).fit(has_constant=False)

# Print the summary of the regression model
print(model3.summary())
```
