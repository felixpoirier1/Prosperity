### Import Statements

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Load Data

The codes below separately load all csv files as a dataframes, rename / expand columns, and drop the original column (where everything is separated by semicolons in a single column). A column 'day' has also been added to all 3 trades dataframes. 

```python
prices_0 = pd.read_csv("prices_round_1_day_0.csv")
prices_0[['day', 'timestamp', 'product', 'bid_price_1', 'bid_volume_1', 'bid_price_2', 'bid_volume_2', 'bid_price_3',
          'bid_volume_3', 'ask_price_1', 'ask_volume_1', 'ask_price_2', 'ask_volume_2', 'ask_price_3', 'ask_volume_3',
          'mid_price', 'profit_and_loss']] = prices_0['day;timestamp;product;bid_price_1;bid_volume_1;bid_price_2;bid_volume_2;bid_price_3;bid_volume_3;ask_price_1;ask_volume_1;ask_price_2;ask_volume_2;ask_price_3;ask_volume_3;mid_price;profit_and_loss'].str.split(';', expand=True)
prices_0.drop(columns=['day;timestamp;product;bid_price_1;bid_volume_1;bid_price_2;bid_volume_2;bid_price_3;bid_volume_3;ask_price_1;ask_volume_1;ask_price_2;ask_volume_2;ask_price_3;ask_volume_3;mid_price;profit_and_loss'], inplace=True)
prices_0
```

```python
prices_1 = pd.read_csv("prices_round_1_day_-1.csv")
prices_1[['day', 'timestamp', 'product', 'bid_price_1', 'bid_volume_1', 'bid_price_2', 'bid_volume_2', 'bid_price_3',
          'bid_volume_3', 'ask_price_1', 'ask_volume_1', 'ask_price_2', 'ask_volume_2', 'ask_price_3', 'ask_volume_3',
          'mid_price', 'profit_and_loss']] = prices_1['day;timestamp;product;bid_price_1;bid_volume_1;bid_price_2;bid_volume_2;bid_price_3;bid_volume_3;ask_price_1;ask_volume_1;ask_price_2;ask_volume_2;ask_price_3;ask_volume_3;mid_price;profit_and_loss'].str.split(';', expand=True)
prices_1.drop(columns=['day;timestamp;product;bid_price_1;bid_volume_1;bid_price_2;bid_volume_2;bid_price_3;bid_volume_3;ask_price_1;ask_volume_1;ask_price_2;ask_volume_2;ask_price_3;ask_volume_3;mid_price;profit_and_loss'], inplace=True)
```

```python
prices_2 = pd.read_csv("prices_round_1_day_-2.csv")
prices_2[['day', 'timestamp', 'product', 'bid_price_1', 'bid_volume_1', 'bid_price_2', 'bid_volume_2', 'bid_price_3',
          'bid_volume_3', 'ask_price_1', 'ask_volume_1', 'ask_price_2', 'ask_volume_2', 'ask_price_3', 'ask_volume_3',
          'mid_price', 'profit_and_loss']] = prices_2['day;timestamp;product;bid_price_1;bid_volume_1;bid_price_2;bid_volume_2;bid_price_3;bid_volume_3;ask_price_1;ask_volume_1;ask_price_2;ask_volume_2;ask_price_3;ask_volume_3;mid_price;profit_and_loss'].str.split(';', expand=True)
prices_2.drop(columns=['day;timestamp;product;bid_price_1;bid_volume_1;bid_price_2;bid_volume_2;bid_price_3;bid_volume_3;ask_price_1;ask_volume_1;ask_price_2;ask_volume_2;ask_price_3;ask_volume_3;mid_price;profit_and_loss'], inplace=True)
```

```python
trades_0 = pd.read_csv('trades_round_1_day_0_nn.csv')
trades_0[['timestamp', 'buyer', 'seller', 'symbol', 'currency', 'price', 'quantity']] = trades_0['timestamp;buyer;seller;symbol;currency;price;quantity'].str.split(';', expand=True)
trades_0.drop(columns=['timestamp;buyer;seller;symbol;currency;price;quantity'], inplace=True)
trades_0.insert(0, 'day', '0')
trades_0
```

```python
trades_1 = pd.read_csv('trades_round_1_day_-1_nn.csv')
trades_1[['timestamp', 'buyer', 'seller', 'symbol', 'currency', 'price', 'quantity']] = trades_1['timestamp;buyer;seller;symbol;currency;price;quantity'].str.split(';', expand=True)
trades_1.drop(columns=['timestamp;buyer;seller;symbol;currency;price;quantity'], inplace=True)
trades_1.insert(0, 'day', '-1')
```

```python
trades_2 = pd.read_csv('trades_round_1_day_-2_nn.csv')
trades_2[['timestamp', 'buyer', 'seller', 'symbol', 'currency', 'price', 'quantity']] = trades_2['timestamp;buyer;seller;symbol;currency;price;quantity'].str.split(';', expand=True)
trades_2.drop(columns=['timestamp;buyer;seller;symbol;currency;price;quantity'], inplace=True)
trades_2.insert(0, 'day', '-2')
```

### Organize Data

The codes below first concatinate all 3 dataframes for price and trades into a single dataframe and then create separate dataframes for amethysts and starfruit respectively. Indexes are also reset. Additionally, columns are converted from object to numeric data types.

```python
prices = pd.concat([prices_2, prices_1, prices_0], ignore_index=True)

prices_columns_to_convert = ['day', 'timestamp', 'bid_price_1', 'bid_volume_1', 'bid_price_2', 'bid_volume_2', 'bid_price_3',
          'bid_volume_3', 'ask_price_1', 'ask_volume_1', 'ask_price_2', 'ask_volume_2', 'ask_price_3', 'ask_volume_3',
          'mid_price', 'profit_and_loss']
for col in prices_columns_to_convert:
    prices[col] = pd.to_numeric(prices[col], errors='coerce')
    
prices
```

```python
amethysts_prices = prices[prices['product'] == 'AMETHYSTS']
amethysts_prices.reset_index(drop=True, inplace=True)
amethysts_prices
```

```python
starfruit_prices = prices[prices['product'] == 'STARFRUIT']
starfruit_prices.reset_index(drop=True, inplace=True)
starfruit_prices
```

```python
trades = pd.concat([trades_2, trades_1, trades_0], ignore_index=True)

trades_columns_to_convert = ['day', 'timestamp', 'price', 'quantity']
for col in trades_columns_to_convert:
    trades[col] = pd.to_numeric(trades[col], errors='coerce')
    
trades
```

```python
amethysts_trades = trades[trades['symbol'] == 'AMETHYSTS']
amethysts_trades.reset_index(drop=True, inplace=True)
amethysts_trades
```

```python
starfruit_trades = trades[trades['symbol'] == 'STARFRUIT']
starfruit_trades.reset_index(drop=True, inplace=True)
starfruit_trades
```
