## PnL Compare

### Import Statments

```python
import pandas as pd
import numpy as np
```

### Load Data

```python
quick_sell = pd.read_csv('Quick_Sell.py.csv')
quick_sell[['day', 'timestamp', 'product', 'bid_price_1', 'bid_volume_1', 'bid_price_2', 'bid_volume_2', 'bid_price_3',
          'bid_volume_3', 'ask_price_1', 'ask_volume_1', 'ask_price_2', 'ask_volume_2', 'ask_price_3', 'ask_volume_3',
          'mid_price', 'profit_and_loss']] = quick_sell['day;timestamp;product;bid_price_1;bid_volume_1;bid_price_2;bid_volume_2;bid_price_3;bid_volume_3;ask_price_1;ask_volume_1;ask_price_2;ask_volume_2;ask_price_3;ask_volume_3;mid_price;profit_and_loss'].str.split(';', expand=True)
quick_sell.drop(columns=['day;timestamp;product;bid_price_1;bid_volume_1;bid_price_2;bid_volume_2;bid_price_3;bid_volume_3;ask_price_1;ask_volume_1;ask_price_2;ask_volume_2;ask_price_3;ask_volume_3;mid_price;profit_and_loss'], inplace=True)
quick_sell = quick_sell[quick_sell['product'] == 'STARFRUIT']
quick_sell['profit_and_loss'] = pd.to_numeric(quick_sell['profit_and_loss'], errors='coerce')
quick_sell['Profit_and_loss_diff'] = quick_sell['profit_and_loss'].diff()
quick_sell
```

```python
penalty = pd.read_csv('Penalty.py.csv')
penalty[['day', 'timestamp', 'product', 'bid_price_1', 'bid_volume_1', 'bid_price_2', 'bid_volume_2', 'bid_price_3',
          'bid_volume_3', 'ask_price_1', 'ask_volume_1', 'ask_price_2', 'ask_volume_2', 'ask_price_3', 'ask_volume_3',
          'mid_price', 'profit_and_loss']] = penalty['day;timestamp;product;bid_price_1;bid_volume_1;bid_price_2;bid_volume_2;bid_price_3;bid_volume_3;ask_price_1;ask_volume_1;ask_price_2;ask_volume_2;ask_price_3;ask_volume_3;mid_price;profit_and_loss'].str.split(';', expand=True)
penalty.drop(columns=['day;timestamp;product;bid_price_1;bid_volume_1;bid_price_2;bid_volume_2;bid_price_3;bid_volume_3;ask_price_1;ask_volume_1;ask_price_2;ask_volume_2;ask_price_3;ask_volume_3;mid_price;profit_and_loss'], inplace=True)
penalty = penalty[penalty['product'] == 'STARFRUIT']
penalty['profit_and_loss'] = pd.to_numeric(penalty['profit_and_loss'], errors='coerce')
penalty['Profit_and_loss_diff'] = penalty['profit_and_loss'].diff()
penalty
```

```python
combined_df = pd.merge(quick_sell[['timestamp', 'Profit_and_loss_diff']], penalty[['timestamp', 'Profit_and_loss_diff']], on='timestamp', suffixes=('_quick_sell', '_penalty'))
combined_df.rename(columns={'Profit_and_loss_diff_quick_sell': 'quick_sell_diff', 'Profit_and_loss_diff_penalty': 'penalty_diff'}, inplace=True)
combined_df['check'] = (combined_df['quick_sell_diff'] == combined_df['penalty_diff']).astype(int)
combined_df
```

```python
total_check_sum = combined_df['check'].sum()

print("Total sum of 'check' column:", total_check_sum)
```
