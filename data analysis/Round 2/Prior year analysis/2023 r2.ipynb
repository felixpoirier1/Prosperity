```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
r2d0 = pd.read_csv('../../../data2023/island-data-bottle-round-2/prices_round_2_day_0.csv')
r2d0[['day', 'timestamp', 'product', 'bid_price_1', 'bid_volume_1', 'bid_price_2', 'bid_volume_2', 'bid_price_3',
          'bid_volume_3', 'ask_price_1', 'ask_volume_1', 'ask_price_2', 'ask_volume_2', 'ask_price_3', 'ask_volume_3',
          'mid_price', 'profit_and_loss']] = r2d0['day;timestamp;product;bid_price_1;bid_volume_1;bid_price_2;bid_volume_2;bid_price_3;bid_volume_3;ask_price_1;ask_volume_1;ask_price_2;ask_volume_2;ask_price_3;ask_volume_3;mid_price;profit_and_loss'].str.split(';', expand=True)
r2d0.drop(columns=['day;timestamp;product;bid_price_1;bid_volume_1;bid_price_2;bid_volume_2;bid_price_3;bid_volume_3;ask_price_1;ask_volume_1;ask_price_2;ask_volume_2;ask_price_3;ask_volume_3;mid_price;profit_and_loss'], inplace=True)
r2d0 = r2d0[r2d0['product'].isin(['COCONUTS', 'PINA_COLADAS'])]
r2d0
```

```python
r2d0['mid_price'] = pd.to_numeric(r2d0['mid_price'], errors='coerce')
r2d0['timestamp'] = pd.to_numeric(r2d0['timestamp'], errors='coerce')
coconuts = r2d0[r2d0['product'] == 'COCONUTS']['mid_price'].mean()
pina_coladas = r2d0[r2d0['product'] == 'PINA_COLADAS']['mid_price'].mean()
pina_coladas / coconuts
```

```python
r2d0_pina = r2d0[r2d0['product'] == 'PINA_COLADAS'].set_index('timestamp')
r2d0_coconut = r2d0[r2d0['product'] == 'COCONUTS'].set_index('timestamp')

df_merged = r2d0_pina[['mid_price']].rename(columns={'mid_price': 'pina_mid_price'}).join(r2d0_coconut[['mid_price']].rename(columns={'mid_price': 'coconut_mid_price'}))

df_merged.dropna(subset=['coconut_mid_price'], inplace=True)

X = df_merged[['coconut_mid_price']]  # Features
y = df_merged['pina_mid_price']  # Target

model = LinearRegression()
model.fit(X, y)

# Extract the coefficient(s) and the intercept
coef = model.coef_[0]  # This extracts the coefficient for 'coconut_mid_price'
intercept = model.intercept_

print("Coefficient for 'coconut_mid_price':", coef)
print("Intercept:", intercept)
```
