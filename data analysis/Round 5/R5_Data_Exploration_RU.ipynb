### Organize Data

```python
r1_data = pd.concat([r1_day_neg2, r1_day_neg1, r1_day_0], ignore_index=True)

trades_columns_to_convert = ['day', 'timestamp', 'price', 'quantity', 'new_timestamp']
for col in trades_columns_to_convert:
    r1_data[col] = pd.to_numeric(r1_data[col], errors='coerce')
    
r1_data
```

```python
r3_data = pd.concat([r3_day_0, r3_day_1, r3_day_2], ignore_index=True)

trades_columns_to_convert = ['day', 'timestamp', 'price', 'quantity', 'new_timestamp']
for col in trades_columns_to_convert:
    r3_data[col] = pd.to_numeric(r3_data[col], errors='coerce')
    
r3_data
```

```python
r3_data.buyer.unique()
```

```python
r3_data.seller.unique()
```

```python
r4_data = pd.concat([r4_day_1, r4_day_2, r4_day_3], ignore_index=True)

trades_columns_to_convert = ['day', 'timestamp', 'price', 'quantity', 'new_timestamp']
for col in trades_columns_to_convert:
    r4_data[col] = pd.to_numeric(r4_data[col], errors='coerce')
    
r4_data
```

```python
r4_data.buyer.unique()
```

### Combine by Person

```python
dfs = [r1_data, r3_data, r4_data]

unique_buyers = set()

for df in dfs:
    unique_buyers.update(df['buyer'].unique())

for buyer in unique_buyers:
    buyer_dfs = [df[df['buyer'] == buyer] for df in dfs]
    buyer_df = pd.concat(buyer_dfs)
    buyer_df = buyer_df.sort_values(by=['day', 'new_timestamp'])
    globals()[f"{buyer}_buyer_df"] = buyer_df
```

```python
dfs = [r1_data, r3_data, r4_data]

unique_sellers = set()

for df in dfs:
    unique_sellers.update(df['seller'].unique())

for seller in unique_sellers:
    seller_dfs = [df[df['seller'] == seller] for df in dfs]
    seller_df = pd.concat(seller_dfs)
    seller_df = seller_df.sort_values(by=['day', 'new_timestamp'])
    globals()[f"{seller}_seller_df"] = seller_df
```

```python
dfs = [r1_data, r3_data, r4_data]

unique_buyers = set()
unique_sellers = set()

for df in dfs:
    unique_buyers.update(df['buyer'].unique())
    unique_sellers.update(df['seller'].unique())

unique_buyers_list = list(unique_buyers)
unique_sellers_list = list(unique_sellers)
```
